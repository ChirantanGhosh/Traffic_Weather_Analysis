# Date-Time splitting

Splitting the date column into month, day and year columns 

```python
def splitDatetime(data) :

   datatime = pd.DatetimeIndex(data.date)  ## Splitting Date-Time column into year, month, day, hour
   data['year'] = datatime.year
   data['month'] = datatime.month
   data['day'] = datatime.day
   data['hour'] = datatime.hour
   return data

splitDatetime(df)
```

# Weather Data Scraping

1. Weather data is scraped from the specified website.
2. Then it's parsed.

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://www.wunderground.com/history/airport/KSFO/2017/1/1/CustomHistory.html?dayend=29&monthend=12&yearend=2017&req_city=&req_state=&req_statename=&reqdb.zip=&reqdb.magic=&reqdb.wmo=")
soup = BeautifulSoup(page.content, 'html.parser')
weather = soup.find(id="observations_details")
table = weather
```

```python
def parse_table(table):
    """ Get data from table """
    return [[cell.get_text().strip() for cell in row.find_all(['th', 'td'])]
           for row in table.find_all('tr')]
```

```python
h= parse_table(table)
results_df = pd.DataFrame.from_records(h)   
```

# Selecting rows and Columns

At this stage only the rows with proper values are selected removing the irrevalent rows with irrevalent values. e.g some rows contain values like '2017', 'avg' etc. which are irrevalent inside 'day' column in the dataframe. This step is also filtering out all those month that are not present in the Traffic data set. At last it is merging out everything and reindexing the rows. 

```python
df2=df1.iloc[2:33,0:] 
df3=df1.iloc[98:128,0:]
df4=df1.iloc[130:161,0:]
df5= df1.iloc[163:193,0:]
df6= df1.iloc[228:259,0:]
df7= df1.iloc[261:291,0:]
df8= df1.iloc[293:324,0:]
df9= df1.iloc[326:356,0:]
df10= df1.iloc[358:387,0:]


fr=[df2,df3,df4,df5,df6,df7,df8,df9,df10]
r=pd.concat(fr)
k = r.reset_index(drop=True)
```

# Month Column creation

1. At this stage a new column is being created for month and proper values are set by comparing this column with the day column.
e.g the day column contans values arranged in asceding order on the basis of month. Now the new month column will contain values in context to day column. All the rows of month column from row index 1 to row index 31 will be filled with '1' as it is the month of Jan. Then with value '2' and it will keep on going until all the months values has been entered. 

```python
k.ix[0:30,'mon'] = '1'
k.ix[31:62,'mon'] = '4'
k.ix[61:92,'mon'] = '5'
k.ix[92:122,'mon'] = '6'
k.ix[122:153,'mon'] = '8'
k.ix[153:183,'mon'] = '9'
k.ix[183:214,'mon'] = '10'
k.ix[214:244,'mon'] = '11'
k.ix[244:,'mon'] = '12'
```

# Unique column creation for joining purpose

A new column Period is being created after concatenating the values from month and day respectively.

```python
k['period'] = k['mon'].astype(str) + k['day'].astype(str)
k['period']= k.period.astype(int)
```

# Merging of 2 Dataframes

At this step the two dataframes created from traffic and Weather data sets are joined together using Left-join and by using unique column 'Period'. 

```python
j=pd.merge(dmf, k, on='period', how='left')
j
```

```python
## Data Type Conversion

j['pdid']= j.pdid.astype(float)
j['temp']= j.temp.astype(float)
j['humi']= j.humi.astype(float)
j['wind']= j.wind.astype(float)
j['preci']= j.preci.astype(float)
j['dew']= j.dew.astype(float)
j['visi']= j.visi.astype(float)
j['day_y']= j.day_y.astype(float)
j['month']= j.month.astype(float)
```

# Plotting and Visualization of Data

1. Number of different analysis has been made after careful visualization of data
2. The first plot shows that, at the day '22' the most number of traffic violation has occured. 
3. The second plot shows that the month of 'November' has most number of traffic violation.

```python
import matplotlib.pyplot as plt
import numpy as np

dff = j[['pdid','day_y']]

count = dff.groupby('day_y')
totalsum = count.aggregate(np.sum).unstack()
totalsum.plot(kind = 'bar', title = 'violation number per day')
plt.ylabel('count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
dff = j[['pdid','month']]

count = dff.groupby('month')
totalsum = count.aggregate(np.sum).unstack()
totalsum.plot(kind = 'bar', title = 'violation number per month')
plt.ylabel('count')
plt.show()
```
